```python
import requests 
#tell Python that you want to use a library with the import statement.
```

```python
req = requests.get("https://www.forbes.com/top-colleges/")
#use the appropriately named get function to issue a GET request. 
#This is equivalent to typing a URL into your browser and hitting enter.
```

```python
req
#The req object however has a custom implementation that shows the object type (i.e. Response) 
#and the HTTP status number (200 means the request was successful).
```

```python
type(req)
#making sure the object agrees with the value above.
```

```python
dir(req)
#listing all the properties of an object
#properties starting with a single and double underscores are not meant to be called directly
```

```python
page=req.text
page
#assign the value of the text property of this requrest object to a variable
#now we have the text of the web page
```

```python
#BeautifulSoup
#another python library
#implement all the different corner cases for us
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page, 'html.parser')
#parser the contents of the page variable using its HTML parser
#then assigns the result of that to the soup variable
```

```python
soup
```

```python
type(page)
```

```python
type(soup) #allows us to see the HTML content in a nice, indented way
```

```python
print(soup.prettify())
```

```python
soup.title
#we can now reference elements of the HTML in different ways using the dot notation
#can access the elements as if they were properties of the object
```

```python
"title" in dir(soup)
#need to make it clear that title is not a property of the soup object
```
